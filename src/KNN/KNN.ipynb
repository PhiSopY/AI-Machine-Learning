# 2. 手写代码实现KNN

```python
from numpy import *
import pandas as pd
import operator
import matplotlib
import matplotlib.pyplot as plt
```

```python
# 读取并处理数据
def file2matrix(filename, featureNum):
    fr = open(filename)
    arrayOLines = fr.readlines()
    numberOfLines = len(arrayOLines)
    returnMat = zeros((numberOfLines,featureNum))
    classLabelVector = []
    index = 0
    for line in arrayOLines:
        line = line.strip()  #截取掉所有的回车字符
        listFromLine = line.split('\t')  #根据'\t'进行分割，成为元素列表
        returnMat[index,:] = listFromLine[1:featureNum+1]
        classLabelVector.append(int(listFromLine[-1]))
        index += 1
    return returnMat, classLabelVector
```

```python
# 函数autoNorm:归一化数值
def autoNorm(dataSet):
    minVals = dataSet.min(0)
    maxVals = dataSet.max(0)
    ranges = maxVals - minVals
    normDataSet = zeros(shape(dataSet))
    m = dataSet.shape[0]  #获取数据行数
    normDataSet = dataSet - tile(minVals, (m,1))
    normDataSet = normDataSet/tile(ranges, (m,1))
    return normDataSet, ranges, minVals
```

```python
# 函数myKNN: KNN核心代码
def myKNN(inX, dataSet, labels, k):
    '''
      inX：用来分类的测试集
      dataSet：用来训练的数据集
      labels：用来训练的数据集的标签
      k：k近邻的k值
    '''
    #欧式距离计算
    dataSetSize = dataSet.shape[0]
    diffMat = tile(inX, (dataSetSize,1)) - dataSet  # tile()：将矩阵inX重复，行方向上重复dataSetSize次，列方向上重复1次
    sqDiffMat = diffMat**2
    sqDistinces = sqDiffMat.sum(axis=1)  #行方向求和
    distances = sqDistinces**0.5
    sortedDistIndicies = distances.argsort() #将距离值进行排序，并返回索引值
    
    #选择距离最小的k个点，并统计k个点中各个类别的数目
    classCount = {}
    for i in range(k):
        voteIlabel = labels[sortedDistIndicies[i]]
        classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1  #get(key,default)：若键key在字典中，返回key的值，否则返回默认值
    
    #排序，选择类别数目最多的所属类别
    sortedClassCount = sorted(classCount.items(), key=operator.itemgetter(1), reverse=True)
    
    return sortedClassCount[0][0]
```

```python
# 函数:dataClassTest: 划分训练集/测试集数据进行测试
def dataClassTest(X, Y, k, testRatio):
    m = X.shape[0]
    numTestVecs = int(m*testRatio)  #得到测试集个数
    errorCount = 0  #记录错误的个数
    for i in range(numTestVecs):
        classifierResult = myKNN(X[i,:], X[numTestVecs:m,:], Y[numTestVecs:m], k)
        print("预测结果：%d， 真实标签：%d"%(classifierResult, Y[i]))
        if(classifierResult != Y[i]):
            errorCount += 1
    print("测试集准确率Acc为：%f"%(1-errorCount/float(numTestVecs)))
```

```python
# 读取特征和标签，分别赋给X,Y
dataPath = '../../data/studentsData.txt'
X, Y = file2matrix(dataPath, 4)
print(X.shape)
print(len(Y))
```

```python
# 数据归一化
X, X_ranges, X_minVals = autoNorm(X)
```

```python
# 绘制分类结果
fig = plt.figure()
plt.scatter(X[:,0],Y, c='b', alpha = 1, s = 100, marker = '+')
plt.scatter(X[:,1],Y, c='m', alpha = 0.5, s = 50, marker = '*')
plt.scatter(X[:,2],Y, c='g', alpha = 0.5, s = 200, marker = 's')
plt.scatter(X[:,3],Y, c='r', alpha = 0.5, s = 150, marker = 'o')
plt.show()
```

```python
# 测试输出准确率
dataClassTest(X,Y,3,0.3)
```

# 3. 调库实现KNN

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
```

```python
def mySklearnKNN(X, Y, k, testRatio):
    ratio = 0.1
    m = X.shape[0]
    numTest = int(m*ratio)
    train_mat = X[numTest:m,:]
    test_mat = X[0:numTest,:]
    train_labels = Y[numTest:m]
    test_labels = Y[0:numTest]
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(train_mat, train_labels)
    print('测试集准确率Acc为: %f'%knn.score(test_mat, test_labels))
```

```python
# 利用Pandas读取特征和标签，分别赋给X,Y
X, Y = file2matrix('studentsData.txt', 4)
```

```python
# 数据归一化
X, X_ranges, X_minVals = autoNorm(X)
```

```python
# 绘制分类结果
fig = plt.figure()
plt.scatter(X[:,0],Y, c='b', alpha = 1, s = 100, marker = '+')
plt.scatter(X[:,1],Y, c='m', alpha = 0.5, s = 50, marker = '*')
plt.scatter(X[:,2],Y, c='g', alpha = 0.5, s = 200, marker = 's')
plt.scatter(X[:,3],Y, c='r', alpha = 0.5, s = 150, marker = 'o')
plt.show()
```

```python
# 测试输出准确率
mySklearnKNN(X,Y,3,0.3)
```
